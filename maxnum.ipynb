```python
import time # Imports the time library for intentional delay.

def find_biggest_number(num1, num2, num3):
  biggest = num1

  if num2 > biggest:
    biggest = num2

  if num3 > biggest:
    biggest = num3

  return biggest
print("Welcome.")
time.sleep(1.0)
print("Please enter three numbers to find the biggest one.")

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

result = find_biggest_number(a, b, c)

print("The biggest number is", result)


print("Exiting system...")
time.sleep(1.0)
print("System exited")

```
